```python
import pandas as pd
import torch
import h3
import json
```

```python
df = pd.read_parquet("../df_la_test.parquet")
```

```python
with open("zone_dfs/labels_dict.json") as f:
    labels_dict = json.load(f)
```

```python
df.loc[:, "h3_index"] = df.apply(lambda row: h3.latlng_to_cell(row["lat"], row["lng"], res=6), axis=1)
df.loc[:, "cluster"] = df.apply(lambda row: labels_dict.get(row["h3_index"], None), axis=1)
```

```python
grouped = df.groupby(by="cluster")
```

```python
for cluster_id, cluster in grouped:
    grouped2 = cluster.groupby(by="route_id")["stop_id"].count()
    max_count = grouped2.max()

    print(f"Cluster {cluster_id} has a maximum length of {max_count}")
```

```python
for cluster_id, cluster in grouped:
    n_routes = cluster["route_id"].nunique()

    if n_routes < 10:
        print(f"Cluster {cluster_id} has {n_routes} routes")
```
